```python
class node:
  def __init__(self,data):
    self.data=data
    self.nref=None
    self.pref=None
class linkedlist:
  def __init__(self):
    self.head=None
  def disf(self):
    if self.head is None:
      print("linked list is empty")
    else:
      n=self.head
      while n is not None:
        print(n.data,"-->",end=" ")
        n=n.nref
  def disr(self):
    if self.head is None:
      print(f"linked list is empty")
    else:
      n=self.head
      while n is not None:
        n=n.pref
  def insert_beg(self,data):
    n=node(data)
    temp=self.head
    temp.pref=n
    n.nref=temp
    self.head=n


  def insert_empty(self,data):
    n_node=node(data)
    if self.head==None:
      self.head=n_node
    else:
      n_node.nref=self.head
      self.head.pref=n_node
      self.head=n_node

  def ine(self,data):
    n_node=node(data)
    if self.head is None:
      print("linked list is empty")
    else:
       n=self.head
       while n.nref is not None:
        n=n.nref
       n.nref=n_node
       n_node.pref=n


  def inspos(self,pos,data):
    n=node(data)
    temp=self.head
    for i in range (1,pos-1):
      temp=temp.nref
    n.pref=temp
    n.nref=temp.nref
    temp.nref.pref=n
    temp.nref=n


  def delb(self):
    if self.head== None:
      print("linked list is empty")
      return
    if self.head.nref is None:
      self.head=None
      print("linked list is empty after deletion")
    else:
      self.head=self.head.nref
      self.head.pref=None

  def delend(self):
    temp=self.head.nref
    before=self.head
    while temp.nref is None:
      temp=temp.nref
      before=before.nref
    before.nref=None
    temp.pref=None




  # def del_p(self,pos):
  #   temp =self.head.nref
  #   b=self.head
  #   for i in range (0,pos-1):
  #     temp=temp.nref
  #     b=b.nref
  #   b.nref = temp.nref
  #   temp.nref.pref=b
  #   temp.nref=None
  #   temp.pref=None
  def del_p(self, pos):
      if self.head is None:
          print("Linked list is empty")
          return
      if pos == 1:
          self.delb()
          return
      temp = self.head
      for i in range(1, pos):
          temp = temp.nref
          if temp is None:
              print("Position out of bounds")
              return
      if temp.nref is None:  # If deleting the last node
          temp.pref.nref = None
      else:
          temp.pref.nref = temp.nref
          temp.nref.pref = temp.pref

L=linkedlist()
n=node(10)
L.head=n
n2=node(20)
n.nref=n2
n2.pref=n
L.insert_beg(5)
L.ine(30)
L.inspos(2,7)
L.delb()
L.delend()

L.del_p(2)
L.disf()
L.disr()
```
